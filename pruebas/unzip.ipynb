```python
import zipfile

def unzip_files(source, destination):
    if source.split('.')[-1].lower() == 'zip':
        # Create a ZipFile object with the path of the zip file
        zip_file = zipfile.ZipFile(source)
        # Extract all the files to a folder
        zip_file.extractall(destination)
        # Close the zip file
        zip_file.close()
```

```python
src = '/home/jaimebarranco/Desktop/test_inference/images/subs_dcm_nii.zip'
dst = '/mnt/sda1/Repos/a-eye/a-eye_segmentation/deep_learning/nnUNet/nnUNet/nnUNet_inference/temp_inference/input'

unzip_files(src, dst)
```
