```python
import subprocess

def print_console(text=None):
    output = subprocess.check_output(f'printf {text}', shell=True)
    output = output.decode('utf-8')
    return output

def print_and_log(text=None, level='info'):
    print(text)
    print_console(text)

command = f' \
        docker run --rm \
        --gpus device=0 \
        --shm-size=gb \
        -v  \
        petermcgor/nnunet:0.0.1 nnUNet_predict \
        -i  \
        -o  \
        -tr nnUNetTrainerV2 \
        -ctr nnUNetTrainerV2CascadeFullRes \
        -m 3d_fullres \
        -p nnUNetPlansv2.1 \
        -t Task313_Eye \
    '
```

```python
print_and_log(f'[A-eye]: nnUNet inference command: {command}', 'info')
```
